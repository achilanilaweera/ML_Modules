```python
#Import Libraries
import os, os.path
import numpy as np
import cv2
import dlib
from imutils import face_utils
```

```python

face_detector=dlib.get_frontal_face_detector()
landmark_detector=dlib.shape_predictor('shape_predictor_68_face_landmarks.dat')

#Import Dataset
data_path=r'E:\SLIIT\Research\Eye datasets\croppedSleepDataSet'
categories=os.listdir(data_path)
print(categories)

category_dict=dict(zip(categories,np.arange(len(categories))))
print(category_dict)
```

```python
#Initializing Variables
window_size=10
eyes=[]
data=[]
target=[]
count=0
max_frames=1000

for category in categories:
    
    category_path=os.path.join(data_path,category)
    video_file_names=os.listdir(category_path)
    
    for video_file_name in video_file_names:
        
        video_file_path=os.path.join(category_path,video_file_name)
        print(video_file_path)
        source=cv2.VideoCapture(video_file_path)
        
        while(True):
            
            ret,img=source.read()
            if(ret==False):
                break
                
            gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
            faces=face_detector(gray)

            for face in faces:

                x1=face.left()
                y1=face.top()
                x2=face.right()
                y2=face.bottom()

                cv2.rectangle(img,(x1,y1),(x2,y2),(0,255,0),2)

                points=landmark_detector(gray,face)
                
                #getting the 68 points of the face
                points=face_utils.shape_to_np(points)
                
                p1=points[36][0]
                p2=points[37][1]
                p3=points[38][1]
                p4=points[39][0]
                p5=points[40][1]
                p6=points[41][1]

                eye_ratio=(abs(p2-p6)+abs(p3-p5))/abs(p1-p4)
                eyes.append(eye_ratio)
                count=count+1
                
                if(count%window_size==0):

                    data.append(eyes)
                    target.append(category_dict[category]) 
                    eyes=[]
                if(count==max_frames):
                    print(count)
                    break
            if(count==max_frames):
                count=0
                break
```

```python

data_new=np.array(data)
target_new=np.array(target)

print(data_new.shape,target_new.shape)
print(data_new[:1])
print(target_new[:1])
```

```python
#Save data to .npy files
np.save('data_train5.npy',data_new)
np.save('target_train5.npy',target_new)
```
