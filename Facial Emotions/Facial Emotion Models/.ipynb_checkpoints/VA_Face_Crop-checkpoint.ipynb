```python
#Import libraries
import cv2
import os

#Cropped the face area in happy image folder
def facecrop(image):  
    facedata = "haarcascade_frontalface_alt.xml"
    cascade = cv2.CascadeClassifier(facedata)

    img = cv2.imread(image)

    try:
    
        minisize = (img.shape[1],img.shape[0])
        miniframe = cv2.resize(img, minisize)

        faces = cascade.detectMultiScale(miniframe)

        for f in faces:
            x, y, w, h = [ v for v in f ]

            sub_face = img[y:y+h, x:x+w]

            f_name = image.split('/')
            f_name = f_name[-1]
            
            print(f_name)
            cv2.imwrite("E:/SLIIT/Research/VA/valence-arousal-recognition-master/valence-arousal-recognition-master/AffectDataFiles/TrainImages/CroppedAffectnetImage/Happy/" + f_name, sub_face)

    except:
        pass


if __name__ == '__main__':
    images = os.listdir("E:/SLIIT/Research/VA/valence-arousal-recognition-master/valence-arousal-recognition-master/AffectDataFiles/TrainImages/train/Happy")
    i = 0
    for img in images:
        file = "E:/SLIIT/Research/VA/valence-arousal-recognition-master/valence-arousal-recognition-master/AffectDataFiles/TrainImages/train/Happy/" + img
        print (i)
        facecrop(file)
        i += 1
```
