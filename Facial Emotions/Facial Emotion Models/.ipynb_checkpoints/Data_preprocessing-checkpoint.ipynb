## Crop Face area

```python
#Import Libraries
import cv2
import os
import numpy as np
from keras.utils import np_utils
```

```python
# cropped faces in angry image folder          
def facecrop(image):  
    facedata = "haarcascade_frontalface_alt.xml"
    cascade = cv2.CascadeClassifier(facedata)

    img = cv2.imread(image)

    try:
    
        minisize = (img.shape[1],img.shape[0])
        miniframe = cv2.resize(img, minisize)

        faces = cascade.detectMultiScale(miniframe)

        for f in faces:
            x, y, w, h = [ v for v in f ]

            sub_face = img[y:y+h, x:x+w]

            f_name = image.split('/')
            f_name = f_name[-1]
            
            print(f_name)
            cv2.imwrite("E:/SLIIT/Research/FER Datasets/Cropped/Angry/" + f_name, sub_face)

    except:
        pass


if __name__ == '__main__':
    images = os.listdir("E:/SLIIT/Research/FER Datasets/Angry")
    i = 0
    for img in images:
        file = "E:/SLIIT/Research/FER Datasets/Angry/" + img
        print (i)
        facecrop(file)
        i += 1
```

## Save cropped images

```python

data_path=r'E:\SLIIT\Research\FER Datasets\Cropped'

categories=os.listdir(data_path)
print(categories)
labels=np.arange(len(categories))
print(labels)
label_dict=dict(zip(categories,labels))
print(label_dict)
```

```python
data=[]
target=[]

for category in categories:
    
    imgs_path=os.path.join(data_path,category)
    img_names=os.listdir(imgs_path)
    
    for img_name in img_names:
        img_path=os.path.join(imgs_path,img_name)
        try:

            data.append(img_path)
            target.append(label_dict[category])
        
        except Exception as e:
            
            print(e)
            
```

```python
data_new=np.array(data)
target_new=np.array(target)
```

```python
print(data_new.shape)
print(target_new.shape)
```

```python
#test the data
target_new[400]
target_new=np_utils.to_categorical(target_new)
```

```python
np.save('data-memory-save-FER',data_new)
np.save('target-memory-save-FER',target_new)
```
